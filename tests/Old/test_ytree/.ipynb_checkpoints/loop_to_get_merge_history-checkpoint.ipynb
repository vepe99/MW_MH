```python
import ytree
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

from PIL import Image
from IPython.display import display
```

# Plotting the merging tree for the a[0]

```python
a = ytree.load('../data/y_tree_data/ahf_halos/snap_N64L16_000.parameter')
fn = a.save_arbor()
a = ytree.load(fn)
```

```python
p = ytree.TreePlot(a[0], dot_kwargs={'rankdir': 'LR', 'size': '"12,6"'})
p.save('a[0].png')

im = Image.open('a[0].png')
display(im)
```

```python
def my_node(halo):
    prog = list(halo.find_root()['prog', 'uid'])
    if halo['uid'] in prog:
        color = 'red'
    else:
        color = 'black'

    label = \
    """
    id: %d
    mass: %.2e Msun
    z: %.3e 
    """ % (halo['uid'], halo['mass'].to('Msun'), halo['redshift'])
    
    my_kwargs = {"label": label, "fontsize": 8,
                 "shape": "square", "color": color}
    return my_kwargs

p = ytree.TreePlot(a[0], dot_kwargs={'rankdir': "BT"},
                   node_function=my_node)
p.save('tree_custom_node.png')
im = Image.open('tree_custom_node.png')
display(im)
```

# Function to get the merging object and track their orbit before the merge

```python
#we can extract the main progenitors in the arbor a[0], this are the red square in the plot before
progenitor_arbor_root_to_leaves = list(a[0]['prog'])
print(progenitor_arbor_root_to_leaves)
```

```python
#in our analysis is easier to go leaves -> root, so we 
progenitor_arbor_leaves_to_root = progenitor_arbor_root_to_leaves[::-1]

#lest's find the first node for which there is a merge
for i in range(len(progenitor_arbor_leaves_to_root)):
    if ( (len(list(progenitor_arbor_leaves_to_root[i]['tree'])) - len(list(progenitor_arbor_leaves_to_root[i-1]['tree'])) )   > 1):
        print('index:', i, 'node:', progenitor_arbor_leaves_to_root[i])
        break
```

```python
#we want to get the first node that interacct with the progenitor line
merged_candidate = []

for i in range(len(progenitor_arbor_leaves_to_root)):
    if ( (len(list(progenitor_arbor_leaves_to_root[i]['tree'])) - len(list(progenitor_arbor_leaves_to_root[i-1]['tree']))) > 1 ):
        l = [j for j in progenitor_arbor_leaves_to_root[i]['tree'] if j['halo_id'] not in list(progenitor_arbor_leaves_to_root[i-1]['tree', 'halo_id']) ]
        merged_candidate.append(l[::-1]) #invert the list to have always leaves -> root
```

```python
oldest_merge = merged_candidate[0]
progenitor_oldest_merge  = progenitor_arbor_leaves_to_root[42-len(oldest_merge):42] 
x = []
y = []
z = []
redshift = []
for i in range(len(oldest_merge)):
    x.append(oldest_merge[i]['position_x'] - progenitor_oldest_merge[i]['position_x']) 
    y.append(oldest_merge[i]['position_y'] - progenitor_oldest_merge[i]['position_y'])
    z.append(oldest_merge[i]['position_z'] - progenitor_oldest_merge[i]['position_z'])
    redshift.append(oldest_merge[i]['redshift']) 
df = pd.DataFrame(data=(np.array(x), np.array(y), np.array(z), np.array(redshift))).T
df.columns = (['x', 'y', 'z', 'redshift'])
```

```python
df
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(df['redshift'], np.sqrt((df['x'])**2 + (df['y'])**2 + (df['z'])**2),  'o--')
ax.set_ylabel('distance from main halo [kpc]')
ax.set_xlabel('redshift')
ax.invert_xaxis()
ax.grid(linestyle='dotted')
```
